```python
from langdetect import detect_langs

result = detect_langs(X_train.iloc[0]['designation'])
#print(type(result[0]))
#help(langdetect.language.Language)
#print(result.lang)
#print(result.prob)

def identify_language(sentence):
    result = detect_langs(sentence)
    lan=[]
    for lang in result:
        lan.append([lang.lang, lang.prob])
    return lan[0][0], lan[0][1], len(lan)

print(identify_language(X_train.iloc[0]['designation']))
```
