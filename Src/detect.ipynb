```python
from bs4 import BeautifulSoup
import lxml
from unidecode import unidecode

def remove_html_tags(text):
    if pd.notna(text):
        soup = BeautifulSoup(text, 'html.parser')
        return soup.getText()
    return ""

def remove_accent(text):
    return unidecode(text)

text = X_train.iloc[4]['description']
print(text)
res = remove_html_tags(text)
print("res = ", res)
res2 = remove_accent(res)
print("res2 = ", remove_special_char(res2))
```

```python
def clean_text(text):
    if pd.isna(text):
        return str('')
    soup = BeautifulSoup(text, 'html.parser')
    tmp = soup.getText()
    tmp = unidecode(tmp)
    return re.sub(r'[^a-zA-Z0-9]+', ' ', tmp)
```

```python
X_train = df_test

text = X_train.iloc[10]['designation']
#text = "Ceci est un essai"
print(text)
print(type(text))

import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)


stop_words_fr = stopwords.words('french')
stop_words_gr = stopwords.words('german')
stop_words_en = stopwords.words('english')
stop_words = stop_words_fr + stop_words_gr + stop_words_en
#print(stop_words)

words = word_tokenize(text) #, language='french')
print(words)
#import fr_dep_news_trf
#nlp = fr_dep_news_trf.load()

lemmatizer = WordNetLemmatizer()

def filter_stopwords(list_words):
    tokens = []
    for word in list_words:
        if word not in stop_words:
            tokens.append(word)
    return tokens

tokens = filter_stopwords(words)

def filter_digits(words):
    tokens = []
    for word in words:
        if not word.isdigit():
            tokens.append(word)
    return tokens

def filter_remove_shortwords(words, threshold):
    tokens = [word for word in words if len(word)>threshold]
    return tokens

tokens = filter_digits(tokens)
print("filtering out digits = ", tokens)

tokens = filter_remove_shortwords(tokens, 2) 
print("Filtering out short words = ", tokens)

def lemmatize(words):
    tokens = []
    for word in words:
        token = lemmatizer.lemmatize(word)
        if token not in tokens:
            tokens.append(token)
    return tokens

print("Lemmatizing = ", lemmatize(tokens))
```
