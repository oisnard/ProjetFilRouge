```python
list_cat_code = df['prdtypecode'].unique()
num_samples = 10

for cat_code in list_cat_code:
    tmp = df.loc[df['prdtypecode']==cat_code].head(num_samples)
    print("Category code = ", cat_code)
    display(tmp[['designation', 'description']])
    lib.show_images_from_id(cat_code, num_samples, 'train')
```

```python
list_cat_code = df['prdtypecode'].unique()
num_samples = 5

print(list_cat_code)

cat_code = 10

def show_firt_cat_infos(df, cat_code, num_samples):
    df_tmp = df.loc[df['prdtypecode']==cat_code].head(num_samples)

    for designation, description, product_id, image_id in zip(df_tmp['designation'].to_list(),
                                                              df_tmp['description'].to_list(),
                                                              df_tmp['productid'].to_list(),
                                                              df_tmp['imageid'].to_list()):
        print("Category ID = ", cat_code, " - product id = ", product_id, " - image id = ", image_id)
        print("   ", designation, "  -   ", description)
        img = lib.get_image(product_id, image_id, 'train')
        if len(img)>0:
            plt.imshow(img)
            plt.axis('off')
            plt.show()

show_firt_cat_infos(df, cat_code, num_samples)



#for cat_code in list_cat_code:
```
